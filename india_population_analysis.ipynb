```python
df['maleliteracy_rate'] = pd.to_numeric(df['maleliteracy_rate'], errors='coerce')
df['female_literacy_rate'] = pd.to_numeric(df['female_literacy_rate'], errors='coerce')
df['average_literacy_rate'] = pd.to_numeric(df['average_literacy_rate'], errors='coerce')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='states/uts', y='average_literacy_rate')
plt.xticks(rotation=90)
plt.title('Average Literacy Rate by State/UT')
plt.show()
```

```python
df[['hindu', 'muslim', 'christian', 'sikhs', 'buddhist', 'others']].head()
```

```python
religion_cols = ['hindu', 'muslim', 'christian', 'sikhs', 'buddhist', 'others']
for col in religion_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.set_index('states/uts')[religion_cols].plot(kind='bar', stacked=True, figsize=(14,7))
plt.title('Religion Distribution by State/UT')
plt.ylabel('Population Percentage')
plt.xticks(rotation=90)
plt.show()
```

```python
df['sex_ratio_number_of_female_per_male'] = pd.to_numeric(df['sex_ratio_number_of_female_per_male'], errors='coerce')

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='states/uts', y='sex_ratio_number_of_female_per_male')
plt.xticks(rotation=90)
plt.title('Sex Ratio (Number of Female per 1000 Male) by State/UT')
plt.show()
```

```python
gdp_cols = ['gdp', 'per_capita_in']
for col in gdp_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
top_gdp = df[['states/uts', 'gdp']].sort_values(by='gdp', ascending=False).head(5)
sns.barplot(data=top_gdp, x='states/uts', y='gdp')
plt.title('Top 5 States/UTs by GDP')
plt.xticks(rotation=45)
plt.show()
```

```python
top_percapita = df[['states/uts', 'per_capita_in']].sort_values(by='per_capita_in', ascending=False).head(5)
sns.barplot(data=top_percapita, x='states/uts', y='per_capita_in')
plt.title('Top 5 States/UTs by Per Capita Income')
plt.xticks(rotation=45)
plt.show()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('india_data.db')

# Table list check karo
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print(tables)

# Table ka first 5 rows dekh lo
df_sql = pd.read_sql("SELECT * FROM state_data LIMIT 5;", conn)
print(df_sql)

conn.close()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agar df_sql se kaam karna hai
plt.figure(figsize=(12,6))
sns.barplot(data=df_sql, x='states/uts', y='population2024')
plt.xticks(rotation=90)
plt.title("Population 2024 by States/UTs")
plt.show()
```
